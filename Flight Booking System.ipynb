```python
def searchByKey(key):
    #key is Flight id
    d=pd.DataFrame(flight)
    d=d.transpose()
    d.columns = [ 'Origin', 'Destination', 'Model', 'Capacity', 'Price']
    print(d.loc[key])
```

```python
def SearchByDestinationOrigin(Dest,Org):
    d=pd.DataFrame(flight)
    d=d.transpose()
    d.columns = [ 'Origin', 'Destination', 'Model', 'Capacity', 'Price']
    print(d[(d.Origin==Org)&(d.Destination==Dest)])
    time.sleep(5)
```

```python
def FlightSeatAllocator(FlightID):
    #print(FlightID)
    cap=flight[FlightID][-2]
    l=dict()
    for i in range(1,cap//4+1):
        for j in "ABCD":
            if len(str(i))==1:
                l["0"+str(i)+j]="NB"
            else:
                l[str(i)+j]="NB"
    flightSeatAllocation[FlightID]=l
```

```python
def Allocation():
    #seat allocation
    for i in flight.keys():
        FlightSeatAllocator(i)
    #mapping booked seats
    for i in range(len(usersFlight['Flight_id'])):
        BookedTicketsMap[usersFlight['Flight_id'][i]]=usersFlight['Seat_number'][i]

    for i in range(len(usersFlight['Flight_id'])):
        flightSeatAllocation[usersFlight['Flight_id'][i]][usersFlight['Seat_number'][i]]='BO'
   
```

```python
def seatMatrix(FlightID):
    c=0
    print()
    print(" "*35+" ->bo : for booked tickets".upper())
    print(" "*35+" ->nb : for not booked tickets".upper())
    print()
    lis=list(flightSeatAllocation[FlightID].keys())
    print(" "*10+" -"*39)
    for i in range(0,len(lis)):
        if c==0:
            print(" "*10,end="")
        print(" # # "+lis[i]+":"+flightSeatAllocation[FlightID][lis[i]]+" # # ",end="  ")
        c+=1
        if c==2:
            print(" "*10,end="")
        if c==4:
            print()
            print(" "*10+" -"*39)
            c=0
```

```python
def GenerateTicket(ID):
    ClearScreen()
    displayHeader()
    print("\n Welcome to Ticket booking portal")
    time.sleep(2)
    name=input(" Enter name : ")
    origin=input(" Enter Origin : ")
    destination=input(" Enter Destination:")
    date=input(" Enter Which Date you wanna travel [YYYY/MM/DD] : ")
    ClearScreen()
    displayHeader()
    time.sleep(2)
    print(" Searching flights From "+origin+" to "+destination)
    time.sleep(2)

    ClearScreen()
    displayHeader()
    print("\n The below are flights available\n")
    SearchByDestinationOrigin(destination,origin)
    flightNum=input("\n Enter Flight Number : ")
    print("\n The below is the matrix display for the fight you have opted".upper())
    seatMatrix(flightNum)
    Seatnumber=input("\n Enter seat number you wanna book : ".upper())
    #usersFlight[name]=[flightNum,name,date,origin,destination,Seatnumber]
    usersFlight['User_id']=ID
    usersFlight['Name'].append(name)
    usersFlight["Travel Date"].append(date)
    usersFlight['Origin']=origin
    usersFlight['Destination'].append(destination)
    usersFlight['Flight_id'].append(flightNum)
    usersFlight['Seat_number'].append(Seatnumber)
    flightSeatAllocation[flightNum][Seatnumber]="BO"
    confirm(flightNum)
```

```python
def createUser():
    ClearScreen()
    displayHeader()
    n=int(input(" Enter number of new users wanna Add : "))
    names=[]
    userids=[]
    passwords=[]
    for i in range(n):
        names.append(input(" Enter name : "))
        userids.append(input(" Enter User Name : "))
        passwords.append(input(" Enter Password : "))
    for i in range(n):
        users[userids[i]]= passwords[i]
```

```python
def createFlight():
    ClearScreen()
    displayHeader()
    n=int(input(" Enter number of new Flights wanna Add : "))
    FlightID=[]
    Forigin=[]
    FDestination=[]
    Model=[]
    Capacity=[]
    price=[]
    for i in range(n):
        FlightID.append(input(" Enter Flight ID : "))
        Forigin.append(input(" Enter origin : "))
        FDestination.append(input(" Enter Destination : "))
        Model.append(input(" Enter Model: "))
        Capacity.append(int(input(" Enter Capacity : ")))
        price.append(int(input(" Enter Price of the travel : ")))
    for i in range(n):
        flight[FlightID[i]]=[Forigin[i],FDestination[i],Model[i],Capacity[i],price[i]]
```

```python
def FlightOccupenceyRate(FlightId):
    df=pd.DataFrame(usersFlight)
    ClearScreen()
    displayHeader()
    print(" Checking for the flight occupancey with id "+FlightId.upper())
    time.sleep(2)
    
    res=df[(df.Flight_id==FlightId)]
    occ=res.Seat_number.count()
    totSeats=flight[FlightId][-2]
    ClearScreen()
    displayHeader()
    print()
    print(" Total Occupancey Rate of Flight ",FlightId," is ",occ/totSeats*100,"%")
    time.sleep(5)
```
